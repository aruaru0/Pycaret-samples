```python
from pycaret.datasets import get_data
from pycaret.classification import *
```

# load dataset

```python
# 用意されているデータセット一覧を表示
all = get_data('index')
```

```python
df = get_data('cancer')
len(df)
```

# Simple EDA

```python
import matplotlib.pyplot as plt
```

```python
df.describe()
```

```python
# クラスの割合をチェック
sum(df.Class == 0), sum(df.Class == 1)
```

```python
plt.figure(figsize=(20,40))
for i, col in enumerate(df.columns) :
    plt.subplot(10, 6, i+1)
    h = plt.boxplot(df[col])
    plt.title(col)
plt.show
```

```python
import seaborn as sns
num_cols = df.select_dtypes(include=['int']).columns.tolist()[1:]
num_cols
```

```python
figsize = (4*4, 40)
fig = plt.figure(figsize=figsize)
for idx, col in enumerate(num_cols):
    ax = plt.subplot(14,4, idx + 1)
    sns.kdeplot(
        data=df, hue='Class', fill=True,
        x=col, palette=['#9E3F00', 'red'], legend=False,
        common_norm=False, # 個別にスケールする
    )

    ax.set_ylabel(''); ax.spines['top'].set_visible(False),
    ax.set_xlabel(''); ax.spines['right'].set_visible(False)
    ax.set_title(f'{col}', loc='right',
                 weight='bold', fontsize=20)

fig.suptitle(f'Features vs Target\n\n\n', ha='center',  fontweight='bold', fontsize=21)
fig.legend([1, 0], loc='upper center', bbox_to_anchor=(0.5, 0.96), fontsize=21, ncol=3)
plt.tight_layout()
plt.show()
```

```python
train, test = df[:600], df[600:]
```

# Setup

```python
seed = 42
clf = setup(data=train, target = 'Class', session_id=seed, fix_imbalance=True) #ignore_features = ['Id'], session_id=42, fix_imbalance=True, remove_outliers=False)
```

# Compare Models (1)

```python
best = compare_models()
```

```python
best
```

# Compare Models (2)

```python
best = compare_models(sort='F1', n_select=3)
```

```python
best
```

# Optimize Model

```python
lgb = create_model('lightgbm')
tuned_lgb = tune_model(lgb)
```

# Ensemble Models

```python
blend = blend_models(estimator_list = best)
```

# Finalize

```python
finalized_model = finalize_model(blend)
```

# Predict

```python
pred = predict_model(finalized_model, data=test[test.columns[1:]])
pred
```

```python
from sklearn.metrics import *
```

```python
confusion_matrix(test.Class, pred.prediction_label)
```

```python
accuracy_score(test.Class, pred.prediction_label)
```

```python
predict_model(finalized_model, data=test[test.columns[1:]], raw_score=True)
```

# Save/Load Models

```python
save_model(finalized_model, "models")
loaded = load_model("models")
```
